* **Data Type Mismatches:** All columns were initially loaded as strings (`object` types).
* **Placeholder Values:** The dataset contained non-standard missing values such as `'UNKNOWN'`, `'ERROR'`, and `'N/A'`.
* **Missing Data:** Significant gaps were found in `Payment Method` (~31%) and `Location` (~39%).
* **Logical Errors:** Calculations for `Total Spent` did not always align with `Quantity * Price Per Unit`.

## 3. Cleaning Methodology
The following steps were taken to ensure data quality:

### A. Standardization & Type Casting
* Converted placeholder strings (`ERROR`, `UNKNOWN`) to `NaN` using `df.replace()`.
* Cast `Quantity`, `Price Per Unit`, and `Total Spent` to **float** types.
* Converted `Transaction Date` to **datetime** objects to allow for time-series analysis.

```python
df = df.replace(['UNKNOWN', 'ERROR', 'N/A', 'nan', 'NaN'], np.nan)

print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
```

```python
print(df['Item'].unique())
print(df['Payment Method'].unique())
```

```python
print(df[['Quantity', 'Price Per Unit']].describe())
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
print(df['Transaction Date'].head())

df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
```

### C. Logical Verification
* Created a `Math_Check` column to validate that $Quantity \times Price = Total Spent$.
* Cleaned the logical inconsistencies to ensure financial reporting accuracy.

```python
df['Math_Check'] = df['Quantity'] * df['Price Per Unit']
errors = df[df['Math_Check'] != df['Total Spent']]
print(f"Logic errors found: {len(errors)}")
```

```python
df.info()
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].median())
df['Total Spent'] = df['Total Spent'].fillna(df['Total Spent'].median())
```

```python
print(df.isnull().sum())
```

```python
df['Location'] = df['Location'].fillna('Unknown')

item_mode = df['Item'].mode()[0]  # Finds the most common item
df['Item'] = df['Item'].fillna(item_mode)

print(df.isnull().sum())
```

```python
df['Payment Method'] = df['Payment Method'].fillna('Unknown')

df = df.dropna(subset=['Transaction Date'])

print(df.isnull().sum())
```

## 4. Visual Analysis
*The following section explores the cleaned data to find trends in item popularity, revenue by location, and payment preferences.*

In this section, we transition from data cleaning to data storytelling. We aim to answer the following research questions to help the Cafe owner optimize operations:

1. **Product Performance:** Which items generate the highest total revenue, and does high volume always equal high profit?
2. **Operational Efficiency:** How do sales compare between 'In-store', 'Takeaway', and 'Online' (Unknown) locations? 
3. **Temporal Patterns:** Are there specific days of the week where the cafe experiences "peak" traffic?

```python
# Prepare Data for Research Questions
df['Day_of_Week'] = df['Transaction Date'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.figure(figsize=(16, 12))

# RQ1: Revenue vs Volume (Scatter/Bubble Plot)
plt.subplot(2, 2, 1)
rq1_data = df.groupby('Item').agg({'Quantity': 'sum', 'Total Spent': 'sum'}).reset_index()
sns.scatterplot(data=rq1_data, x='Quantity', y='Total Spent', size='Total Spent', 
                hue='Item', sizes=(100, 1000), palette='deep')
plt.title('RQ1: Revenue vs. Volume by Item')
plt.grid(True, linestyle='--', alpha=0.6)

# RQ2: Location Performance (Boxplot for Spending Habits)
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='Location', y='Total Spent', hue='Location', palette='Set2', legend=False)
plt.title('RQ2: Distribution of Spending by Location')
plt.ylim(0, df['Total Spent'].quantile(0.95)) # Focus on the bulk of data, ignoring extreme outliers

# RQ3: Daily Traffic (Countplot)
plt.subplot(2, 1, 2)
sns.countplot(data=df, x='Day_of_Week', order=day_order, hue='Day_of_Week', palette='coolwarm', legend=False)
plt.title('RQ3: Transaction Volume by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Transactions')

plt.tight_layout()
plt.show()

# Print a quick summary table for the user
print("--- Quick Insights Summary ---")
print(f"Busiest Day: {df['Day_of_Week'].mode()[0]}")
print(f"Highest Revenue Item: {rq1_data.loc[rq1_data['Total Spent'].idxmax(), 'Item']}")
```
