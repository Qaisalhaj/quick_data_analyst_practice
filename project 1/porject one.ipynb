# Practice Data Cleaning: "Dirty" Cafe Sales

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### after importing the important libraries now move to import the dataset

```python
df = pd.read_csv('dirty_cafe_sales.csv')
df.head()
```

### here we check the datttypes for the dataset and we find alot of columns treated as string thats a validity problem 

```python
df.info()
```

### here we check the missing values

```python
print(df.isnull().sum())
```

### here we checked the unique values to make sure the missing values enterd correctly

```python
print(df['Payment Method'].unique())
```

```python
df = df.replace(['UNKNOWN', 'ERROR', 'N/A', 'nan', 'NaN'], np.nan)

print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
```

```python
print(df['Item'].unique())
print(df['Payment Method'].unique())
```

```python
print(df[['Quantity', 'Price Per Unit']].describe())
```

```python
df['Math_Check'] = df['Quantity'] * df['Price Per Unit']
errors = df[df['Math_Check'] != df['Total Spent']]
print(f"Logic errors found: {len(errors)}")
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
print(df['Transaction Date'].head())

df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
```

```python
df.info()
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].median())
df['Total Spent'] = df['Total Spent'].fillna(df['Total Spent'].median())

```

```python
print(df.isnull().sum())
```

```python
df['Location'] = df['Location'].fillna('Unknown')

item_mode = df['Item'].mode()[0]  # Finds the most common item
df['Item'] = df['Item'].fillna(item_mode)

print(df.isnull().sum())
```

```python
df['Payment Method'] = df['Payment Method'].fillna('Unknown')

df = df.dropna(subset=['Transaction Date'])

print(df.isnull().sum())
```
